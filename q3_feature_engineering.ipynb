```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Load Data
df = pd.read_csv('/Users/umeshkapil/Downloads/q3_retail_promotions.csv')

# 2. Feature Engineering
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

df['year'] = df['transaction_date'].dt.year
df['month'] = df['transaction_date'].dt.month
df['day_of_week'] = df['transaction_date'].dt.dayofweek
df['is_month_end'] = (df['transaction_date'].dt.day >= 25).astype(int)

print(df.head())

# 3. Time-based Split
df = df.sort_values('transaction_date')

train = df.iloc[:int(0.8 * len(df))]
test = df.iloc[int(0.8 * len(df)):]

X_train = train.drop('items_sold', axis=1)
y_train = train['items_sold']

X_test = test.drop('items_sold', axis=1)
y_test = test['items_sold']

# 4. Columns
num_cols = ['competition_density']
cat_cols = ['promotion_type', 'location_type', 'store_size']

# 5. Pipeline
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(), cat_cols)
])

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

# 6. Train
pipeline.fit(X_train, y_train)

# 7. Predict
y_pred = pipeline.predict(X_test)

# 8. Evaluate
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
